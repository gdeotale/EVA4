```
import matplotlib.pylab as plt
import numpy as np
import argparse

VERSION = '0.1'

def factor(shape, h, i):
    if shape == "triangle":
        return (1/(h*h) * (-1)**i)
    else:
        return (1/h)
		
f = 1
odd = 0.5
mult = 1
    
t = np.linspace(0, 8, num=8000)
y = np.zeros(8000)
    
# compute fundamental and each harmonic
for i in range(int(1000)+1):
    h = i * mult + odd
    yh = factor('triangle',h,i) * np.sin(h * 2 * np.pi * f * t)
    y = y+yh

y = abs(y)
    
plt.plot(t, y)
plt.show()
```
